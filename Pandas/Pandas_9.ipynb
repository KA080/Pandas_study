```python
df['range'] = df['high'] - df['low']
df.head()
```

```python
"""컬럼 삭제
-df.drop('컬럼명', axis=1)
    -2차원 배열이기에 컬럼이나 로우를 선택해서 삭제할 수 있음
    -원본은 그대로 유지되고 컬럼이 삭제된 새로운 데이터프레임 객체가 리턴됨
    (inplace를 쓰면 원본값으로 만들수 있음)
"""
df.drop('value', axis=1)
```

```python
"""시계열 데이터와 인덱스
    -시계열 데이터는 인덱스가 날짜와 시간으로 구성됨
        -문자열로 표현된 날짜와 시간을 DatetimeIndex 타입으로 변환해서 사용해야함
"""
import pandas as pd

date = ['2023-11-24','2023-11-25']
dt = pd.to_datetime(date)

print(type(date))
print(type(dt))
print(type(dt[0]))
```
